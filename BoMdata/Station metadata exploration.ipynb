### Add information on physical location of stations:

```r
# Load in station lookup table from README page of water regs .xls file
metadata_waterregs_stations <- read_csv('Station_Metadata_FinalIFD_Water Regulations_Continuous_README_page.csv', skip = 18, col_names = c("Agency", "OrganisationName", "State"))
# Remove blank rows and values
metadata_waterregs_stations <- metadata_waterregs_stations %>% drop_na()
# Lok at the DF:
metadata_waterregs_stations
```

# Zooming in on our Sydney domain!

Setting the domain to be smaller only works somewhat... 
If we try to plot stations within a bounded region, the plot will only extend as far as the outer stations' location. 
On the other hand, we want to plot the whole region to the boundary, so we set that with limits within `scale_x_continuous`.

```r
# Where are my domain bounds?
syd_domain <- 
c(left = 148.7, right = 153.2, top = -31.9, bottom = -35.9)
```

```r
# Map without domain adjustment:
syd_map <- ggmap(get_map(syd_domain, maptype = "toner-lite", zoom = 8), 
                 base_layer = ggplot(data = nsw_data, aes(Longitude, Latitude)))
syd_map +
geom_point(aes(col = OrganisationName), alpha = 0.6)
```

```r
# Map WITH adjustment:
syd_map +
    scale_x_continuous(limits = c(syd_domain['left'], syd_domain['right']), expand = c(0, 0)) +
    scale_y_continuous(limits = c(syd_domain['bottom'], syd_domain['top']), expand = c(0, 0)) +
    geom_point(aes(col = OrganisationName), alpha = 0.6)
```

Hmmmm... Not much different!!!! :(

## Testing different length station records:

```r
# Histogram of station durations (from first year and last year)
nsw_data %>%
  ggplot(aes(x = Number.of.Years, fill = OrganisationName)) +
  geom_histogram(binwidth = 1) +
  guides(fill = guide_legend(ncol = 2)) +
  #theme_classic() +
  theme(legend.position="bottom",legend.direction="vertical") +
  labs(fill = "Organisation", x = "Length of Station Record (years)")
```

So most stations have less than 30 years of data. :(

```r
nsw_data %>% 
  group_by(Number.of.Years) %>% 
    summarise(num = n()) %>%
    arrange(desc(Number.of.Years)) %>%
    mutate(numExceedingNumberOfYears = cumsum(num)) 
```

```r
nsw_data %>% 
  group_by(Number.of.Years) %>% 
    summarise(num = n()) %>%
    arrange(desc(Number.of.Years)) %>%
    mutate(numExceedingNumberOfYears = cumsum(num), percentExceeding = cumsum(num)*100/sum(num)) %>%
    filter(Number.of.Years == 30 | Number.of.Years == 25 | Number.of.Years == 20) %>%
    select(Number.of.Years, percentExceeding)
```

So, from a total of 412 stations, we have just over half of stations with 20+ years; 37% with 25+ years, and <b>only 13.5% with 30+ years</b>.
